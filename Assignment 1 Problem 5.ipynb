Matrix inverse calc

```python
M1_inv = np.linalg.inv(M1)
print("M1 inverse =")
print(M1_inv)
```

```python
print("M1 × M1⁻¹ =")
print(M1 @ M1_inv)
```

# Problem 5: Average, Variance, and Correlation

The dataset for this assignment is available [here](https://drive.google.com/drive/folders/1FV7qofHD_olqygrIjASLQ7k7LTyVSEGG?usp=sharing). The prediction task is to predict the price of a house (column price) given the other features. Please ignore the columns id and date, as well as the categorical column zipcode. File kc_house_data.csv includes all the records in the dataset. You should use the entire dataset for the assignment. You can also find a Word document including the feature description in the same folder. In this problem, we will perform some exploratory data analysis using the house price dataset.

```python
import pandas as pd
import numpy as np
```

```python
# Load the data
df = pd.read_csv('kc_house_data.csv')

print(f"Dataset shape: {df.shape[0]} rows, {df.shape[1]} columns")
df.head()
```

```python
# Features to analyze (excluding id, date, zipcode, and price which is the response)
exclude_cols = ['id', 'date', 'zipcode', 'price']
feature_cols = [col for col in df.columns if col not in exclude_cols]

print(f"Features to analyze ({len(feature_cols)} total):")
print(feature_cols)
```

---
## Part 1: Compute Statistics for Each Feature

1. For each feature, write code to compute the average value, the min and max values, as well as its variance.
    
    Which features have the lowest and the highest average? Include the feature name and their average values for the features with the
    lowest and highest average.
    
    Which features have the lowest and the highest variance? Include the feature name and their variance values for the features with the
    lowest and highest variance.

```python
# Compute statistics for each feature
stats = pd.DataFrame({
    'Min': df[feature_cols].min(),
    'Max': df[feature_cols].max(),
    'Average (Mean)': df[feature_cols].mean(),
    'Variance': df[feature_cols].var()
})

print("Full Statistics Table:")
stats
```

### Which features have the lowest and highest average?

```python
# Find features with lowest and highest average
min_avg_feature = stats['Average (Mean)'].idxmin()
max_avg_feature = stats['Average (Mean)'].idxmax()
min_avg_value = stats['Average (Mean)'].min()
max_avg_value = stats['Average (Mean)'].max()

print(f"Feature with LOWEST average:  {min_avg_feature} = {min_avg_value:.6f}")
print(f"Feature with HIGHEST average: {max_avg_feature} = {max_avg_value:.6f}")
```

### Which features have the lowest and highest variance?

```python
# Find features with lowest and highest variance
min_var_feature = stats['Variance'].idxmin()
max_var_feature = stats['Variance'].idxmax()
min_var_value = stats['Variance'].min()
max_var_value = stats['Variance'].max()

print(f"Feature with LOWEST variance:  {min_var_feature} = {min_var_value:.6f}")
print(f"Feature with HIGHEST variance: {max_var_feature} = {max_var_value:.6f}")
```

---
## Part 2: Correlation Coefficient with the Response

2. Compute the correlation coefficient of each feature with the response. Include a table with the correlation coefficient of each feature with the response. 
Which features are positively correlated (i.e., have positive correlation coefficient) with the response?

Which feature has the highest positive correlation with the response?

```python
# Compute correlation of each feature with price
correlations = df[feature_cols].corrwith(df['price'])

# Create a sorted dataframe for display
corr_df = pd.DataFrame({
    'Feature': correlations.index,
    'Correlation with Price': correlations.values
}).sort_values('Correlation with Price', ascending=False)

print("Correlation Table (sorted by correlation):")
corr_df
```

### Which features are positively correlated with the response?

```python
# Identify positively correlated features
positive_corr = correlations[correlations > 0]

print(f"Features with POSITIVE correlation ({len(positive_corr)} total):")
for feat, corr in positive_corr.sort_values(ascending=False).items():
    print(f"  {feat}: +{corr:.6f}")
```

### Which feature has the highest positive correlation with the response?

```python
# Feature with highest positive correlation
max_corr_feature = correlations.idxmax()
max_corr_value = correlations.max()

print(f"Feature with HIGHEST positive correlation: {max_corr_feature} = {max_corr_value:.6f}")
```

---
## Part 3: Features with Negative Correlation

Were you able to find any features with a negative correlation coefficient with the response?

```python
# Check for negatively correlated features
negative_corr = correlations[correlations < 0]

if len(negative_corr) > 0:
    print(f"Yes, {len(negative_corr)} feature(s) have NEGATIVE correlation with price:")
    for feat, corr in negative_corr.sort_values().items():
        print(f"  {feat}: {corr:.6f}")
else:
    print("No features with negative correlation were found.")
    print("All 17 features are positively correlated with price.")
```
